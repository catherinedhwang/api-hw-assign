### Store Part I results into DataFrame

```python
# Import city_data_df from Part 1 as CSV file

new_df = pd.read_csv("../starter_code/city_data_df.csv")
new_df
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Configure gmaps
gmaps.configure(api_key=g_key)
```

```python
# Store latitude/longitude as locations
locations = new_df[['Latitude', 'Longitude']].astype(float)

# Store as humidity_complete
humidity_complete = new_df['Humidity'].astype(float)
```

```python
figure_layout = {
    'width': '900px',
    'height': '600px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}


fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=humidity_complete, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 4)

fig.add_layer(heat_layer)
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# Create dataframe based on ideal weather conditions:
#   A max temperature lower than 80 degrees but higher than 70.

#   Wind speed less than 10 mph.

#   Zero cloudiness.

narrowed_ideal_df = new_df.loc[(new_df['Max Temperature'] > 70) & 
                            (new_df['Max Temperature'] < 80) & 
                            (new_df['Cloudiness'] == 0), :]
# Drop any cities that do not meet the ideal conditions
narrowed_ideal_df = narrowed_ideal_df.dropna()

# Clean up new dataframe
narrowed_ideal_df.reset_index(inplace=True)
del narrowed_ideal_df['index']

narrowed_ideal_df.head()
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# Store variable named hotel_df
hotel_df = narrowed_ideal_df

# Add a "Hotel Name" column to the DF
hotel_df['Hotel Name'] = ''
hotel_df.head()
```

```python
# Set up a parameters dictionary
params = {
    'radius': 5000,
    'types': 'lodging',
    'keyword': 'hotel',
    'key': g_key
}

# base_url
base_url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'

for index, row in hotel_df.iterrows():
    # Get longitude and latitude from hotel_df
    lat = row['Latitude']
    lng = row['Longitude']
    city = row['City']
    
    # Add keyword to params dict
    params['location'] = f'{lat}, {lng}'

    # Assemble url and make API request
    print(f"Retrieving Results for Index {index}: {city}.")
    response = requests.get(base_url, params=params).json()

    # extract results
    results = response['results']
    
    try:
        print(f"Closest hotel in {city} is {results[0]['name']}.")
        
        hotel_df.loc[index, 'Hotel name'] = results[0]['name']
        
    except (KeyError, IndexError):
        print('Missing field/result... skipping.')
    
#     time.sleep(1)
        
    print("------------")    
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Latitude", "Longitude"]]
```
